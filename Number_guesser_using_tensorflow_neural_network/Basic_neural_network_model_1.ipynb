```python
import tensorflow as tf #libraries to import tensor,flow numpy and keras
import numpy as np
from tensorflow import keras
```

```python
my_model = tf.keras.Sequential([keras.layers.Dense(units = 1, input_shape = [1])])
#creating a new sequential neural network with one unit 
```

```python
#now compile the unit with optimizer sgd and loss of mean squared error
my_model.compile( optimizer = 'sgd', loss = 'mean_squared_error')
#  'STOCHASTIC GRADIENT DESCENT' - sgd. 
```

```python
model1 = tf.keras.Sequential([keras.layers.Dense(units = 1, input_shape = [1])])
model1.compile( optimizer = 'sgd', loss = 'mean_squared_error')
xs = np.array([-1,  0, 1, 2, 3, 4], dtype=int)
ys = np.array([-3, -1, 1, 3, 5, 7], dtype=int)
model1.fit(xs, ys, epochs= 500)
print((model1.predict([10])))
x = []
for val in np.array(model1.predict([num for num in range(1,100)])):
    for v in val:
        x.append(int(round(v)))
        print(int(round(v)), end = "\t")
print("\n",x)
```
